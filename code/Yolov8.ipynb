Chia má»™t cÃ¡ch ngáº«u nhiÃªn

```python
import os
import shutil
import random

# ÄÆ°á»ng dáº«n Ä‘áº¿n thÆ° má»¥c chá»©a áº£nh vÃ  nhÃ£n
image_dir = '/content/drive/MyDrive/ColabNotebooks/dataset/images'
label_dir = '/content/drive/MyDrive/ColabNotebooks/dataset/labels'

# ÄÆ°á»ng dáº«n Ä‘áº¿n cÃ¡c thÆ° má»¥c train vÃ  validation
train_image_dir = '/content/drive/MyDrive/ColabNotebooks/train/images'
train_label_dir = '/content/drive/MyDrive/ColabNotebooks/train/labels'
val_image_dir = '/content/drive/MyDrive/ColabNotebooks/val/images'
val_label_dir = '/content/drive/MyDrive/ColabNotebooks/val/labels'

# Táº¡o cÃ¡c thÆ° má»¥c train vÃ  validation náº¿u chÆ°a tá»“n táº¡i
os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

# Danh sÃ¡ch cÃ¡c file áº£nh vÃ  nhÃ£n
images = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.JPG'))]
labels = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

print(f"Sá»‘ lÆ°á»£ng áº£nh: {len(images)}")
print(f"Sá»‘ lÆ°á»£ng nhÃ£n: {len(labels)}")

# Láº¥y tÃªn file khÃ´ng tÃ­nh pháº§n má»Ÿ rá»™ng
image_files = {os.path.splitext(f)[0]: f for f in images}
label_files = {os.path.splitext(f)[0]: f for f in labels}

# TÃ¬m cÃ¡c file áº£nh vÃ  nhÃ£n khá»›p nhau
common_files = set(image_files.keys()) & set(label_files.keys())

# Lá»c láº¡i danh sÃ¡ch áº£nh vÃ  nhÃ£n Ä‘á»ƒ chá»‰ láº¥y cÃ¡c file khá»›p nhau
images = [image_files[f] for f in common_files]
labels = [label_files[f] for f in common_files]

print(f"Sá»‘ lÆ°á»£ng áº£nh sau khi khá»›p: {len(images)}")
print(f"Sá»‘ lÆ°á»£ng nhÃ£n sau khi khá»›p: {len(labels)}")

# Äáº£m báº£o má»—i áº£nh cÃ³ má»™t nhÃ£n tÆ°Æ¡ng á»©ng
assert len(images) == len(labels)

# Trá»™n ngáº«u nhiÃªn danh sÃ¡ch áº£nh vÃ  nhÃ£n
data = list(zip(images, labels))

# Äáº·t seed Ä‘á»ƒ Ä‘áº£m báº£o tÃ­nh nháº¥t quÃ¡n
random.seed(42)
random.shuffle(data)

images, labels = zip(*data)

# TÃ­nh sá»‘ lÆ°á»£ng áº£nh cho train vÃ  validation
train_size = int(0.8 * len(images))  # Tá»· lá»‡ 80-20

# Chia táº­p train vÃ  validation
train_images = images[:train_size]
train_labels = labels[:train_size]
val_images = images[train_size:]
val_labels = labels[train_size:]

# Sao chÃ©p áº£nh vÃ  nhÃ£n vÃ o thÆ° má»¥c train
for img, lbl in zip(train_images, train_labels):
    shutil.copy(os.path.join(image_dir, img), os.path.join(train_image_dir, img))
    shutil.copy(os.path.join(label_dir, lbl), os.path.join(train_label_dir, lbl))

# Sao chÃ©p áº£nh vÃ  nhÃ£n vÃ o thÆ° má»¥c validation
for img, lbl in zip(val_images, val_labels):
    shutil.copy(os.path.join(image_dir, img), os.path.join(val_image_dir, img))
    shutil.copy(os.path.join(label_dir, lbl), os.path.join(val_label_dir, lbl))

print('Chia táº­p dá»¯ liá»‡u thÃ nh cÃ´ng!')
```

Chia cÃ³ chiáº¿n lÆ°á»£c

```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

# ÄÆ°á»ng dáº«n Ä‘áº¿n thÆ° má»¥c chá»©a áº£nh vÃ  nhÃ£n
image_dir = '/content/drive/MyDrive/ColabNotebooks/dataset/images'
label_dir = '/content/drive/MyDrive/ColabNotebooks/dataset/labels'

# ÄÆ°á»ng dáº«n Ä‘áº¿n cÃ¡c thÆ° má»¥c train vÃ  validation
train_image_dir = '/content/drive/MyDrive/ColabNotebooks/train/images'
train_label_dir = '/content/drive/MyDrive/ColabNotebooks/train/labels'
val_image_dir = '/content/drive/MyDrive/ColabNotebooks/val/images'
val_label_dir = '/content/drive/MyDrive/ColabNotebooks/val/labels'

# Táº¡o cÃ¡c thÆ° má»¥c train vÃ  validation náº¿u chÆ°a tá»“n táº¡i
os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

# Danh sÃ¡ch cÃ¡c file áº£nh vÃ  nhÃ£n
images = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.JPG'))]
labels = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

print(f"Sá»‘ lÆ°á»£ng áº£nh: {len(images)}")
print(f"Sá»‘ lÆ°á»£ng nhÃ£n: {len(labels)}")

# Láº¥y tÃªn file khÃ´ng tÃ­nh pháº§n má»Ÿ rá»™ng
image_files = {os.path.splitext(f)[0]: f for f in images}
label_files = {os.path.splitext(f)[0]: f for f in labels}

# TÃ¬m cÃ¡c file áº£nh vÃ  nhÃ£n khá»›p nhau
common_files = set(image_files.keys()) & set(label_files.keys())

# Lá»c láº¡i danh sÃ¡ch áº£nh vÃ  nhÃ£n Ä‘á»ƒ chá»‰ láº¥y cÃ¡c file khá»›p nhau
images = [image_files[f] for f in common_files]
labels = [label_files[f] for f in common_files]

print(f"Sá»‘ lÆ°á»£ng áº£nh sau khi khá»›p: {len(images)}")
print(f"Sá»‘ lÆ°á»£ng nhÃ£n sau khi khá»›p: {len(labels)}")

# Äáº£m báº£o má»—i áº£nh cÃ³ má»™t nhÃ£n tÆ°Æ¡ng á»©ng
assert len(images) == len(labels)

# HÃ m Ä‘á»c nhÃ£n tá»« file nhÃ£n
def read_label(filepath):
    with open(filepath, 'r') as file:
        return file.readline().strip().split()[0]

# Äá»c nhÃ£n tá»« file nhÃ£n
labels_content = [read_label(os.path.join(label_dir, lbl)) for lbl in labels]

# Sá»­ dá»¥ng train_test_split Ä‘á»ƒ chia táº­p dá»¯ liá»‡u vá»›i chiáº¿n lÆ°á»£c chia táº§ng
train_images, val_images, train_labels, val_labels = train_test_split(
    images, labels, test_size=0.2, random_state=42, stratify=labels_content)

# Sao chÃ©p áº£nh vÃ  nhÃ£n vÃ o thÆ° má»¥c train
for img, lbl in zip(train_images, train_labels):
    shutil.copy(os.path.join(image_dir, img), os.path.join(train_image_dir, img))
    shutil.copy(os.path.join(label_dir, lbl), os.path.join(train_label_dir, lbl))

# Sao chÃ©p áº£nh vÃ  nhÃ£n vÃ o thÆ° má»¥c validation
for img, lbl in zip(val_images, val_labels):
    shutil.copy(os.path.join(image_dir, img), os.path.join(val_image_dir, img))
    shutil.copy(os.path.join(label_dir, lbl), os.path.join(val_label_dir, lbl))

print('Chia táº­p dá»¯ liá»‡u thÃ nh cÃ´ng!')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO

# Load the trained model
model = YOLO('C:/Users/ACER/Documents/PBL5/best.pt')

# Evaluate the model on the validation set
metrics = model.val(data='C:/Users/ACER/Documents/PBL5/mydataset.yaml')

# Extract the evaluation metrics
precision = metrics.box.p
recall = metrics.box.r
f1_score = metrics.box.f1

# Print the evaluation metrics for each class
print("Per-class metrics:")
for i, (p, r, f1) in enumerate(zip(precision, recall, f1_score)):
    print(f'Class {i}:')
    print(f'  Precision: {p:.4f}')
    print(f'  Recall: {r:.4f}')
    print(f'  F1 Score: {f1:.4f}')

# Compute and print the average metrics
avg_precision = np.mean(precision)
avg_recall = np.mean(recall)
avg_f1_score = np.mean(f1_score)

print('\nAverage metrics:')
print(f'  Precision: {avg_precision:.4f}')
print(f'  Recall: {avg_recall:.4f}')
print(f'  F1 Score: {avg_f1_score:.4f}')
```

```python
import os
from ultralytics import YOLO
import cv2

# ÄÆ°á»ng dáº«n Ä‘áº¿n mÃ´ hÃ¬nh Ä‘Ã£ huáº¥n luyá»‡n
model_path = 'C:/Users/ACER/Documents/PBL5/best.pt'

# Load mÃ´ hÃ¬nh
model = YOLO(model_path)

# ÄÆ°á»ng dáº«n Ä‘áº¿n thÆ° má»¥c chá»©a táº­p test
test_dir = 'C:/Users/ACER/Documents/PBL5/test'

# ÄÆ°á»ng dáº«n Ä‘á»ƒ lÆ°u káº¿t quáº£ dá»± Ä‘oÃ¡n
output_dir = 'C:/Users/ACER/Documents/PBL5/predictions'
os.makedirs(output_dir, exist_ok=True)

# Dá»± Ä‘oÃ¡n cho táº¥t cáº£ cÃ¡c áº£nh trong thÆ° má»¥c test
for img_name in os.listdir(test_dir):
    img_path = os.path.join(test_dir, img_name)
    
    # Äá»c áº£nh
    img = cv2.imread(img_path)
    
    # Dá»± Ä‘oÃ¡n
    results = model.predict(img)
    
    # LÆ°u káº¿t quáº£ dá»± Ä‘oÃ¡n (hiá»ƒn thá»‹ káº¿t quáº£ trÃªn áº£nh vÃ  lÆ°u áº£nh)
    result_img = results[0].plot()  # Láº¥y káº¿t quáº£ Ä‘áº§u tiÃªn vÃ  váº½ lÃªn áº£nh
    output_path = os.path.join(output_dir, f'pred_{img_name}')
    cv2.imwrite(output_path, result_img)

    # In ra káº¿t quáº£ dá»± Ä‘oÃ¡n (tuá»³ chá»n)
    print(f"Predicted {img_name} and saved to {output_path}")
```
